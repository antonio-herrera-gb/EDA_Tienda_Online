## ANALISIS DE VENTAS DE UNA TIENDA ONLINE

### HIPOTESIS:
- El Q4 fue el trimestre con más ingresos.
- Reino Unido es el país que más nos compra.
- Las Devoluciones son menos del 5% frente a las Ventas.

### Saber mas:
- Cual es el producto más vendido y su proyeccion de venta durante el año.
- Producto más devuelto y queremos verlo por trimestre también.
- Precio medio de compra.
- Que país nos devuelve más.

```python
# Importaciones

import pandas as pd
```

```python
ruta_datos_brutos = "../data/online_retail_II.xlsx"
```

```python
df_datos_brutos = pd.read_excel(ruta_datos_brutos)
```

```python
# Una vision de los datos
df_datos_brutos
```

```python
# Vision del DF
df_datos_brutos.info()
```

```python
# Sabiendo que "Quantity" no tiene nulos, vamos a sacar las columnas "Devolucion" y "Ventas"
df_datos_brutos["Devolucion"] = df_datos_brutos["Quantity"] < 0
df_datos_brutos["Ventas"] = df_datos_brutos["Quantity"] > 0

# Vemos como quedan las nuevas columnas:
df_datos_brutos.sample(5)
```

```python
# Nulos
df_datos_brutos.isnull().sum()
```

```python
# Tras revisar la hipótesis, no creo que sea de gran conocer el Customer ID. Pero si corregir el nombre
df_datos_brutos.rename(columns={'Customer ID': 'Customer_ID'}, inplace=True)
df_datos_brutos.sample(5)
```

```python
# Vemos los nulos de Description
df_datos_brutos[df_datos_brutos.Description.isnull()]
```

```python
# Voy a mirar el StockCode porque tenemos un Nulo en su descripcion 
# pero lo mismo tenemos el mismo StockCode grabado en otra venta y podemos sacar su description.
df_datos_brutos[df_datos_brutos.StockCode == 21324]
```

```python
# Viendo que la teoria funciona con ese nulo puede ser que funcione con los demas. Asi que vamos a intentar rellenar esos nulos.
# Iteramos los valores unicos de StockCode y asignamos el primer valor a Description:
for stock_code, description in df_datos_brutos.groupby("StockCode")["Description"].first().items():
    df_datos_brutos.loc[df_datos_brutos["StockCode"] == stock_code, "Description"] = description
```

```python
# Ya hemos recuperado todos los datos posibles de esa Description.
df_datos_brutos.isnull().sum()
```

```python
# Veamos si ha rellenado bien el nulo que teniamos:
df_datos_brutos[df_datos_brutos.StockCode == 21324]
```

```python
# Si vemos ya esta rellenada la celda, asi que el codigo ha funcionado.
# Los nulos que tenemos de Description ya no seria tan seguro asociarle un valor ya que en una BBDD StockCode y Description son los que van unidos.
df_datos_brutos[df_datos_brutos.Description.isnull()]
```

```python
# Hemos visto que los nulos de Description estan asociados a un Price de 0 euros y a un Customer_ID NaN. Vamos a eliminar esas filas.
df_datos_brutos= df_datos_brutos.dropna(subset=["Description"])
df_datos_brutos.isnull().sum()
```

```python
# Veamos si tenemos articulos de 0 euros.
df_datos_brutos[df_datos_brutos.Price == 0.0]
```

```python
filtro_visualizacion = (df_datos_brutos.StockCode == 21733) & (df_datos_brutos.Devolucion == True)
resultado = df_datos_brutos[filtro_visualizacion]
resultado
```

```python
# Vemos que tenemos la posibilidad de rellenar datos en Price ya que tenemos otros Price. Ademas vemos que ocurre con mas frecuencia si es una Devolucion.
# Hacemos igual que antes. Iteramos pero esta vez nos quedaremos con la media del Price ya que podemos tener precios muy distintos conforme el año.

for stock_code, price in df_datos_brutos.groupby("StockCode")["Price"].mean().items():
    df_datos_brutos.loc[df_datos_brutos["StockCode"] == stock_code, "Price"] = price
```

```python
# Vamos a ver ahora los Precios 0 que teniamos.
df_datos_brutos[df_datos_brutos.Price == 0.0]
```

```python
# Vemos que quedan muy pocas filas, vamos a eliminarlo porque no tenemos ninguna informacion coherente.

df_datos_limpios = df_datos_brutos[df_datos_brutos.Price != 0.0]
df_datos_limpios[df_datos_limpios.Price == 0.0]
df_datos_limpios[df_datos_limpios.Price == 0]
```

```python
# Vamos hacer una nueva columna de Precio_Total. Ya que en un futuro queremos tener los datos calculados tanto de Devolucion como de Ventas.
# Para ello, tenemos un "popurri" en Quantity y es que cuando es Devolucion el valor es negativo. Para no tener problemas lo ponemos todo en positivo.

df_datos_limpios.loc[(df_datos_limpios.Devolucion == True) & (df_datos_limpios.Quantity < 0), "Quantity"] = df_datos_limpios.Quantity.abs()
df_datos_limpios.Quantity.value_counts()
```

```python
# Pensando bien en un futuro analisis, la columna de "Customer_ID" donde tenemos Nulos y hemos dicho que para las Hipotesis no nos importa que sea nulo...
# Vamos a darle un nombre para tenerlo mas estetico:

df_datos_limpios["Customer_ID"] = df_datos_limpios["Customer_ID"].fillna(12345.0)
```

```python
df_datos_limpios
```

```python
# Por ultimo calculemos la nueva columna "Precio_Total" para en el futuro poderla analizar sin tener que hacer muchos calculos.
df_datos_limpios["Precio_Total"] = round(df_datos_limpios["Quantity"] * df_datos_limpios["Price"], 2)
df_datos_limpios
```

```python
df_datos_limpios.isnull().sum()
```
